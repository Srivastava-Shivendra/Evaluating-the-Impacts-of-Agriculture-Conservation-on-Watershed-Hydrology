```python
import pandas as pd
```

```python
df = pd.read_excel("ShellCreek_Precipitation.xlsx", parse_dates=["Date"])
df = df.sort_values("Date")
```

```python
# Define thresholds
thresholds = [20, 30, 40, 50]
periods = { "2004-2013": ("2004-01-01", "2013-12-31"), "2014-2023": ("2014-01-01", "2023-12-31")}
results = {}
for period_name, (start, end) in periods.items():
    mask = (df["Date"] >= start) & (df["Date"] <= end)
    subset = df.loc[mask]
    counts = {f">{thr}mm": (subset["Precipitation"] > thr).sum() for thr in thresholds}
    results[period_name] = counts
    
results_df = pd.DataFrame(results)
print(results_df)
```
